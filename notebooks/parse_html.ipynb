```python
def find_html_files(directory):
    html_files = []
    
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith('.html'):
                html_files.append(os.path.join(root, file))
    
    return html_files

directory = '/Users/evgeny/Downloads/Telegram Desktop'
html_files = find_html_files(directory)
print(len(html_files))
```

```python
example_file2 = "/Users/evgeny/Downloads/Telegram Desktop/ChatExport_2024-09-11 (1)/messages.html"

def find_all_strings_by_file(example_file):
    with open(example_file, 'r', encoding='utf-8') as file:
        html_content = file.read()

    soup = BeautifulSoup(html_content, 'html.parser')

    divs = soup.find_all('div', class_='text')

    messages = [div.get_text() for div in divs]

    return messages
```

```python
emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # —Å–º–∞–π–ª–∏–∫–∏
        "\U0001F300-\U0001F5FF"  # —Å–∏–º–≤–æ–ª—ã –∏ –ø–∏–∫—Ç–æ–≥—Ä–∞–º–º—ã
        "\U0001F680-\U0001F6FF"  # —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç –∏ —Å–∏–º–≤–æ–ª—ã
        "\U0001F700-\U0001F77F"  # –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ —Å–∏–º–≤–æ–ª—ã
        "\U0001F780-\U0001F7FF"  # –≥–µ–æ–º–µ—Ç—Ä–∏—á–µ—Å–∫–∏–µ —Å–∏–º–≤–æ–ª—ã
        "\U0001F800-\U0001F8FF"  # –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ —Å–∏–º–≤–æ–ª—ã
        "\U0001F900-\U0001F9FF"  # –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ —Å–∏–º–≤–æ–ª—ã
        "\U0001FA00-\U0001FA6F"  # –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ —Å–∏–º–≤–æ–ª—ã
        "\U0001FA70-\U0001FAFF"  # –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ —Å–∏–º–≤–æ–ª—ã
        "\U00002702-\U000027B0"  # —Å–∏–º–≤–æ–ª—ã
        "\U000024C2-\U0001F251"
        "]+", flags=re.UNICODE)

def contains_emoji(text):
    return bool(emoji_pattern.search(text))
```

```python
messages = [message for message in messages if contains_emoji(message)]
print(len(messages))
print(messages[6])
```

```python
text = "dfgfdgdgüòäüòäüòä.a–± üòä–ë—É–¥–µ–º —Ä–∞–¥—ã, –µ—Å–ª–∏ –≤–∞–º —É–¥–∞—Å—Ç—Å—è –ø—Ä–∏–Ω—è—Ç—å —É—á–∞—Å—Ç–∏–µ –≤ –Ω–∞—à–∏—Ö —ç–∫–æ-–∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—è—Ö.üòä"

def trunc_text(text):
    res = text.split(".")
    i = len(res) - 1
    while i >= 0: 
        if len(res[i]) > 1:
            text = res[i]
            break
        i -= 1
        if i < 0:
            text = ""
        
    res = text.split()
    text = " ".join(res[-15:])
    return text

def split_string_on_text_and_emoji(message):
        result = []
        while message and emoji_pattern.search(message):
                start, end = emoji_pattern.search(message).span()
                text_message = message[:start].strip()
                text_message =  trunc_text(text_message)
                if len(text_message) > 1:
                        result.append((text_message, message[start:end]))
                message = message[end:]
        return result

print(split_string_on_text_and_emoji(text))
```
